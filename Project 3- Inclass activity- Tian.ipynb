```python
# compute the sum of lst2
# Building a function allows us to reuse the code without rebuilding it.

# In[2]:

# define a function to get the sum of a list
# input: a list
# output: the sum of all the elements in the list
def total(list_obj):
    total = 0
    n = len(list_obj)
    # n is the length of the list
    #for i in range(n):
    #    val = list_obj[i]
    #    total += val
    for val in list_obj:
        #total += val
        total = total + val
    # return outputs whatever value or variable is after return
    return total


# define a function to get the mean value of a list
# input: a list
# output: the mean of all the elements in the list
def mean(list_obj):
    n = len(list_obj)
    # n is the length of the list
    mean_ = total(list_obj) / n
    # mean_ is the mean value of the list
    return mean_

# define a function to get the median value of a list
# input: a list
# output: the median of all the elements in the list
def median(list_obj):
    n = len(list_obj)
    # order list by value from smallest to greatest
    list_obj = sorted(list_obj)
    
    # sort the list by the values first
    # if the list is not even
    if n % 2 != 0:
        middle_index = int((n - 1) / 2)
        median_ = list_obj[middle_index]
    # if the list length is odd, the mean value will be the middle element
    else:
        upper_middle_index = int(n / 2)
        lower_middle_index = upper_middle_index - 1
        # take the average of the two middle numbers
#        median_ = mean([list_obj[lower_middle_index],list_obj[upper_middle_index]])
        median_ = mean(list_obj[lower_middle_index:upper_middle_index + 1])
    # if the list length is even, the mean value will be the average of the two middle elements, with the index of n/2 and n/2-1
    return median_

# define a function to get the mode (the elements with the most frequency) of a list
# input: a list
# output: the mode of all the elements in the list
def mode(list_obj):
    max_count = 0
    # counter dict records # of times each value appears in list
    counter_dict = {}
    # create a map 
    for value in list_obj:
        counter_dict[value] = 0
    for value in list_obj:
        counter_dict[value] += 1
    # count the frequce of each element in the list and store them in the map, as the key will the element and value will be the frequence
    # find maximum number of times a value appears
    
    # sort the frequence
    count_list = list(counter_dict.values())
    
    # get the most highest frequence
    max_count = max(count_list)
    
    # search the map by finding the element with the value equals to max_count
    mode_ = [key for key in counter_dict if counter_dict[key] == max_count]
    return mode_

# define a function to get the variance of a list
# input: a list
# output: the variance of all the elements in the list
# we have two options to compute the variance, one is by dividing by n, another is by dividing by n - 1
def variance(list_obj, sample = False):
    # sampe = False assumes lst represents entire population
    list_mean = mean(list_obj)
    # get the average value of the list by calling the mean function
    n = len(list_obj)
    sum_sq_diff = 0
    for val in list_obj:
        sum_sq_diff += (val - list_mean) ** 2
    # get the square value of the diffrence of each element and the average
    if sample == False:
        variance_ = sum_sq_diff / n
    else:
        variance_ = sum_sq_diff / (n - 1)
    # get the variance value
    return variance_

# define a function to get the SD of a list
# input: a list
# output: the sd of all the elements in the list
def SD(list_obj, sample = False):
    SD_ = variance(list_obj, sample) ** (1/2)
    # get the variance by calling the variance function
    return SD_

# define a function to get the covariance of two lists
# input: two lists
# output: the covariance of two list
# we have two options to compute the covariance, one is by dividing by n, another is by dividing by n - 1
def covariance(list_obj1, list_obj2, sample = False):
    mean1 = mean(list_obj1)
    # get the mean value of list1
    mean2 = mean(list_obj2)
    # get the mean value of list2
    cov = 0
    n1 = len(list_obj1)
    # get the length of list1

    n2 = len(list_obj2)
    # get the length of list1
    if n1 == n2:
        n = n1
        
        for i in range(n1):
            cov += (list_obj1[i] - mean1) * (list_obj2[i] - mean2)
            # get the total convariance of two lists
        if sample == False:
            cov = cov / n
        else:
            cov = cov / (n - 1)
        return cov
    else:
        print("Lists are not of equal length")
        print("Length of List1:", n1)
        print("Length of List2:", n2)

# define a function to get the correlation of two lists
# input: two lists
# output: the correlation of two list
def correlation(list_obj1, list_obj2):
    cov = covariance(list_obj1, list_obj2)
    # ge the covariance of two lists by calling the covariance function
    SD1 = SD(list_obj1)
    # ge the SD of list1 by calling the SD function
    SD2 = SD(list_obj2)
    # ge the SD of list1 by calling the SD function

    corr = cov / (SD1 * SD2)
    
    return corr


# define a function to get the skewness of a list
# input: a list
# output: the skewness of two list
# we have two options to compute the skewness, one is by dividing by n, another is by dividing by n - 1
def skewness(list_obj, sample = False):
    mean_ = mean(list_obj)
    # ge the mean of the list by calling the mean function
    SD_ = SD(list_obj, sample)
    # ge the SD of the list by calling the SD function, sample should be the same as the input option
    skew = 0
    n = len(list_obj)
    for val in list_obj:
        skew += (val - mean_) ** 3
        skew = skew / (n *  SD_ ** 3) if not sample else n * skew / ((n-1) * (n - 2) * SD_ ** 3)
        
    return skew

# define a function to get the kurtosis of a list
# input: a list
# output: the kurtosis of two list
# we have two options to compute the kurtosis, one is by dividing by n, another is by dividing by n - 1
def kurtosis(lst, sample = False):
    mean_ = mean(lst)
    # ge the mean of the list by calling the mean function
    kurt = 0
    SD_ = SD(lst, sample)
    # get the SD of the list by calling the SD function, sample should be the same as the input option
    n = len(lst)
    # get the length of th list
    for x in lst:
        kurt += (x - mean_) ** 4
    # iterate the list to compute the sum of (x - mean_) ** 4
    kurt = kurt / (n * SD_ ** 4) if sample == False else  n * (n + 1) * kurt /     ((n - 1) * (n - 2) * (n - 3) * (SD_ ** 4)) - (3 *(n - 1) ** 2) / ((n - 2) * (n - 3))
    
    return kurt
```
